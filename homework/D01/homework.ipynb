題目:<br>
1.生成一個等差數列，首數為0，尾數為20，公差為1的數列。<br>
2.呈上題，將以上數列取出偶數。<br>
3.呈1題，將數列取出3的倍數。<br>

```python
import numpy as np
```

```python
#1.生成一個等差數列，首數為0，尾數為20，公差為1的數列。
a = np.arange(21)
a
```

```python
#2.呈上題，將以上數列取出偶數。
for i in a: 
    if i % 2 == 0: print(i)
```

```python
#3.呈1題，將數列取出3的倍數。
for i in a:
    if i % 3 == 0: print(i)
```
